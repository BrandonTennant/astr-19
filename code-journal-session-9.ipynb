```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random = np.random.uniform(0, 1, 1000)
```

```python
plt.figure()
plt.hist(random, bins=100, color='skyblue', edgecolor='black')
plt.xlabel("Random Number Value")
plt.ylabel("Frequency")
```

```python
plt.savefig("uniform_distribution_histogram.pdf")
plt.show()
```
