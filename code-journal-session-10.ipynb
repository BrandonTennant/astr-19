```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_numbers = np.random.wald(mean=1, scale=1, size=1000)
```

```python
plt.figure()
plt.hist(random_numbers, bins=100, color='pink', edgecolor='black')
plt.xlabel("Random Number Value")
plt.ylabel("Frequency")
plt.title("Histogram of Wald Distributed Random Numbers")
```

```python
plt.savefig("wald_distribution_histogram.pdf")
plt.show()
```
